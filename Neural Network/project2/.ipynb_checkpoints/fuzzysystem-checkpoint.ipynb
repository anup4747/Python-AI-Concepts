```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define fuzzy sets A and B
A = [(1, 0.3), (2, 0.6), (3, 0.8), (4, 0.9), (5, 1.0)]
B = [(1, 0.4), (2, 0.5), (3, 0.7), (4, 0.8), (5, 0.9)]

# Extract x and membership values for A and B
x_A, membership_A = zip(*A)
x_B, membership_B = zip(*B)

# 1. Fuzzy union (A  B) using the max operation
union_x = list(set(x_A) | set(x_B))  # Combine the unique x values from A and B
union_membership = [max(membership_A[x_A.index(x)] if x in x_A else 0, membership_B[x_B.index(x)] if x in x_B else 0) for x in union_x]

print(union_x)
print(union_membership)
# 2. Fuzzy intersection (A  B) using the min operation
intersection_x = list(set(x_A) & set(x_B))  # Common x values between A and B
intersection_membership = [min(membership_A[x_A.index(x)], membership_B[x_B.index(x)]) for x in intersection_x]

print(intersection_x)
print(intersection_membership)
# 3. Plot the membership functions of A, B, A  B, and A  B

plt.figure(figsize=(10, 6))

# Plot A
plt.plot(x_A, membership_A, label='A', marker='o', linestyle='-', color='blue')

# Plot B
plt.plot(x_B, membership_B, label='B', marker='o', linestyle='-', color='red')

# Plot A  B (fuzzy union)
plt.plot(union_x, union_membership, label='A  B', marker='o', linestyle='--', color='green')

# Plot A  B (fuzzy intersection)
plt.plot(intersection_x, intersection_membership, label='A  B', marker='o', linestyle='--', color='purple')

# Labels and title
plt.title('Fuzzy Set Operations')
plt.xlabel('x')
plt.ylabel('Membership value')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A= [(1,0.2), (2,0.5), (3,0.7), (4,0.9), (5,1.0)]

x_values = [x[0] for x in A]
A_values = [x[1] for x in A]

neg_A_Values = [1 - x for x in A_values]

plt.plot(x_values,A_values, label= "A (original set)", marker= 'o', linestyle= '-', color='b')
plt.plot(x_values,neg_A_Values, label= "A (neg set)", marker= 'o', linestyle= '-', color='r')
 
# Add labels and title
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.title('Fuzzy Set A and Its Complement A')

# Display legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

    
```
