```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]
ex = [0.4,0,0,0]

plt.pie(x, labels = y)
plt.title("Pie Chart")
plt.show()
```

```python

x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]
ex = [0.4,0,0,0]
c = ["r","b","g","y"]
# plt.pie(x, labels = y)
plt.pie(x, labels = y , explode = ex,autopct = "%0.1f%%", colors = c, shadow = True)
plt.title("Pie Chart")
plt.show()
```

```python

x = [10,20,30,40]
y = ["C", "C++", "Java", "Python"]
ex = [0.4,0,0,0]
c = ["r","b","g","y"]
# plt.pie(x, labels = y)
plt.pie(x, labels = y , explode = ex,autopct = "%0.1f%%", colors = c, shadow = True, counterclock = False)
plt.title("Pie Chart")
plt.legend(loc= 1)
plt.show()
```
