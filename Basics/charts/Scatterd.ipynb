```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [2,3,1,4,5,3,6]
colors = ["r", "y","g","b","r","g","r"]
plt.scatter(x,y,c = colors,s = 50)
t = plt.colorbar()
t.set_label("ColorBar", fontsize = 15)
plt.title("Scattered", fontsize=15)
plt.xlabel("Day", fontsize = 15)
plt.ylabel("No." ,fontsize = 15)
plt.show()
```
