```python
import matplotlib.pyplot as plt
x = ["python","c", "c++", "java"]
y = [85,70,60,82]
z= [20,30,40,50]

plt.xlabel("Language" ,fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
plt.bar(x,y,width = 0.4 , label = "popularity")
plt.bar(x,z,width = 0.4 , label = "popularity2")
plt.show()
```

```python
import numpy as np

x = ["python","c", "c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]
print(p)
print(p1)

plt.xlabel("Language" ,fontsize = 15)
plt.ylabel("No.", fontsize = 15)
plt.title("wscube", fontsize = 15)
plt.bar(p,y,width , label = "popularity")
plt.bar(p1,z,width, label = "popularity2")


plt.xticks(p+width/2, x, rotation = 90)
plt.legend()
plt.show()
```
